```python
class Solution:
    def twoSum(self, nums, target):
        dic = {}
        for i, j in enumerate(nums):
            sub = target-j
            if  sub in dic:
                return (dic[sub], i)
            else:
                dic[j] = i
a = Solution()
print(a.twoSum(nums=[3,2,4], target=6))
```
